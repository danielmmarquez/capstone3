# citys realestate price time series forcasting

Covid-19 has has a tremendous impact on all of us both physically and mentally however, the global economy also felt the effects of covid. Intuitively, we may thing the economy would crash however, in the United States the real-estate market has been booming with record breaking housing costs. the point of this project is to predict which city will experience the largest growth in real-estate prices.

# 3. Algorithms & Machine Learning

There where a total of 4 different models that where implemented, 
1. Linear regression <br>
this model was first implemented but was quickly found to not be useful due to the auto-correlation of the time stamps since the data is of type time series. 
2. Exponential smoothing (Exponential weighted Moving Average) <br>
the next model implemented was Exponential smoothing, which was able to make a prediction however, there was an issue when it came to predicting the test set. The main issue was due to the large increase prices the homes where sold in 2020. The model was trained on housing prices from 2017 to 2021 however, there is a large increase of housing prices after the year 2020 meaning that the model did not expect the spike of housing cost there for making its predictions inaccurate. 
3. ARIMA time series forecasting <br>
Auto Regressive Integrated Moving Average was implemented next however, this ARIMA model experienced the same issue which was present in the Exponential soothing model that being, the model was train on data pre-covid or pre-market boom. The Model did not do a good job predicting the test set so it was decided that there was a need for a model that was more complex. 
4. LSTM <br>
Long Short-Term Memory was the model that did a better job at keeping up with the sudden rise in real-estate prices. this recurrent neural network managed to predict the test data set more accurately.
